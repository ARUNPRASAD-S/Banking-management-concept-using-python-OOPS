```python
from bank_accounts import BankAccount

accounts = {}

def create_account():
    name = input("Enter your name: ")
    age = input("Enter your age: ")

    new_account = BankAccount(name, age)
    accounts[new_account.account_number] = new_account

    print(f"\nAccount created successfully!")
    print(f"Your account number is: {new_account.account_number}")

def login():
    acc_num = input("Enter your account number: ").strip()
    account = accounts.get(acc_num)

    if account:
        print(f"\nWelcome back, {account.name}!")
        account_menu(account)
    else:
        print("Account not found. Please check the account number.")

def account_menu(account):
    while True:
        print("\n===== Account Menu =====")
        print("1.Deposit")
        print("2.Withdraw")
        print("3.Check Balance")
        print("4.View Account Info")
        print("5.Logout")

        choice = input("Choose an option (1-5): ").strip()

        if choice == '1':
            try:
                amount = float(input("Enter amount to deposit: ₹"))
                account.deposit(amount)
            except ValueError:
                print("Please enter a valid number.")

        elif choice == '2':
            try:
                amount = float(input("Enter amount to withdraw: ₹"))
                account.withdraw(amount)
            except ValueError:
                print("Please enter a valid number.")

        elif choice == '3':
            account.check_balance()

        elif choice == '4':
            account.display_info()

        elif choice == '5':
            print("Logged out successfully.")
            break

        else:
            print("Invalid option. Try again.")

def main():
    while True:
        print("\n=== Welcome to our bank ===")
        print("1.Create New Account")
        print("2.Login to Existing Account")
        print("3.Exit")
        choice = input("Select an option (1-3): ")
        if choice == '1':
            create_account()
        elif choice == '2':
            login()
        elif choice == '3':
            print("Thank you for using Python Bank.")
            break
        else:
            print("Invalid selection. Please choose 1, 2, or 3.")
main()
```
